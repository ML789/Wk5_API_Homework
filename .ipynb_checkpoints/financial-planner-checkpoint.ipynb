```python
# Format current date as ISO format
# get the date for yesterday, note: today will clash with the market being open, yesterday close price is best to evaluate
date_yesterday = date.today() - timedelta(days = 1)
# convert date_today to iso format with timezone in New York
date_yesterday_iso = pd.Timestamp(str(date_yesterday), tz="America/New_York").isoformat()

# Set the tickers
tickers = ["AGG", "SPY"]

# Set timeframe to "1Day" for Alpaca API
timeframe = "1Day"

# Get current closing prices for SPY and AGG
df_portfolio = alpaca.get_bars(
    tickers,
    timeframe,
    start = date_yesterday_iso,
    end = date_yesterday_iso
).df

df_portfolio

# Reorganize the DataFrame
# Separate ticker data
# YOUR CODE HERE!
df_portfolio = df_portfolio.set_index('symbol')


# Concatenate the ticker DataFrames
# YOUR CODE HERE!
## already completed with the API method, by setting "tickers" to list of strings of the stocks to assess

# Preview DataFrame
df_portfolio.head()
```

```python
# Pick AGG and SPY close prices
# YOUR CODE HERE!
agg_close_price = df_portfolio.loc['AGG']['close']
spy_close_price = df_portfolio.loc['SPY']['close']

# Print AGG and SPY close prices
print(f"Current AGG closing price: ${agg_close_price}")
print(f"Current SPY closing price: ${spy_close_price}")
```

```python
# Compute the current value of shares
# YOUR CODE HERE!
my_agg_value = my_agg *agg_close_price
my_spy_value = my_spy *spy_close_price

# Print current value of shares
print(f"The current value of your {my_spy} SPY shares is ${my_spy_value:0.2f}")
print(f"The current value of your {my_agg} AGG shares is ${my_agg_value:0.2f}")
```

### Savings Health Analysis

```python
# Set monthly household income
# YOUR CODE HERE!
monthly_income = 12000

# Consolidate financial assets data
# YOUR CODE HERE!
df_savings = {'Saving_Value':
              [my_btc_value + my_eth_value,
               my_spy_value + my_agg_value]
             }



# Create savings DataFrame
# YOUR CODE HERE!
df_savings = pd.DataFrame(df_savings, index=['Crypto', 'Shares'])

# Display savings DataFrame
display(df_savings)
```

```python
# Plot savings pie chart
# YOUR CODE HERE!
df_savings.plot.pie(y = 'Saving_Value', figsize = (10,10), title = "Pie Chart of Savings Portfolio" )
```

```python
# Set ideal emergency fund
emergency_fund = monthly_income * 3

# Calculate total amount of savings
# YOUR CODE HERE!
total_savings = df_savings.sum().values

# Validate saving health
# YOUR CODE HERE!

if total_savings > emergency_fund:
    print("Well done, you have more than required in your emergency fund")
elif total_savings == emergency_fund:
    print("Well done, you have met the minimum requirement, keep saving to be safe")
else:
    print(f"Sorry, You need to save ${round(emergency_fund - total_savings,2)} more to have a good emergency fund")
    
```

## Part 2 - Retirement Planning

### Monte Carlo Simulation

```python
# Set start and end dates of five years back from today.
# Sample results may vary from the solution based on the time frame chosen

# use variable "date_yesterday" defined earlier in the code (note due to leap years there is 365.25 days in a year)
start_date = pd.Timestamp(date_yesterday - timedelta(days = 5*365.25), tz='America/New_York').isoformat()
end_date = pd.Timestamp(date_yesterday, tz='America/New_York').isoformat()

```

```python
# Get 5 years' worth of historical data for SPY and AGG
# YOUR CODE HERE!
df_stock_data = alpaca.get_bars(
    tickers,
    timeframe,
    start = start_date,
    end = end_date
).df


# create temporary data storage variable
df_stock_data_dict = {}


# Reorganize the DataFrame
# Separate ticker data
# YOUR CODE HERE!

for symbol in df_stock_data['symbol'].unique():
    df_stock_data_dict[symbol] = df_stock_data[ df_stock_data['symbol'] == symbol ].drop(columns = 'symbol')
    



# Concatenate the ticker DataFrames
# YOUR CODE HERE!
df_stock_data_concat = pd.concat(df_stock_data_dict.values(), axis = 'columns', 
                                 keys = df_stock_data_dict.keys()
                                )



# Display sample data
df_stock_data_concat.head()
```

```python
df_stock_data_concat.shape
```

```python
tickers
```

```python
# Configuring a Monte Carlo simulation to forecast 30 years cumulative returns
# YOUR CODE HERE!
MC_simulation = MCSimulation(
    portfolio_data = df_stock_data_concat,
    weights = [.40,.60],
    num_simulation = 500,
    num_trading_days = 252*30
)
```

```python
# Printing the simulation input data
# YOUR CODE HERE!

MC_simulation.portfolio_data.head()
```

```python
# Running a Monte Carlo simulation to forecast 30 years cumulative returns
# YOUR CODE HERE!
MC_simulation.calc_cumulative_return()

```

```python
# Plot simulation outcomes
# YOUR CODE HERE!
MC_simulation.plot_simulation()
```

```python
# Plot probability distribution and confidence intervals
# YOUR CODE HERE!

MC_simulation.plot_distribution()
```

### Retirement Analysis

```python
# Fetch summary statistics from the Monte Carlo simulation results
# YOUR CODE HERE!
summary = MC_simulation.summarize_cumulative_return()

# Print summary statistics
# YOUR CODE HERE!
print(summary)
```

### Calculate the expected portfolio return at the `95%` lower and upper confidence intervals based on a `$20,000` initial investment.

```python
# Set initial investment
initial_investment = 20000

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $20,000
# YOUR CODE HERE!
ci_lower = round(summary[8]*initial_investment,2)
ci_upper = round(summary[9]*initial_investment,2)

# Print results
print(f"There is a 95% chance that an initial investment of ${initial_investment} in the portfolio\n"
      f"over the next 30 years will end within in the range of: \n"
      f"${ci_lower} and ${ci_upper}")
```

### Calculate the expected portfolio return at the `95%` lower and upper confidence intervals based on a `50%` increase in the initial investment.

```python
# Set initial investment
initial_investment = 20000 * 1.5

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $30,000
# YOUR CODE HERE!
ci_lower = round(summary[8]*initial_investment,2)
ci_upper = round(summary[9]*initial_investment,2)


# Print results
print(f"There is a 95% chance that an initial investment of ${initial_investment} in the portfolio\n"
      f"over the next 30 years will end within in the range of: \n"
      f"${ci_lower} and ${ci_upper}")
```
